```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("train.csv")
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.value_counts()
```

```python
Titanic = titanic.drop_duplicates()
Titanic
```

```python
Titanic.isnull().sum()
```

```python
Titanic = Titanic.fillna(value=0)
Titanic
```

```python
Titanic.isnull().sum()
```

```python
Titanic.value_counts()
```

```python
plt.figure(figsize=(10,8),dpi=50)
sns.barplot(data= Titanic, x='Sex',y='Survived',color='lightgreen')
plt.show()
```

```python
plt.figure(figsize=(8,6),dpi=50)
sns.pairplot(data=Titanic,hue='Sex',palette='Set1')
plt.show();
```

```python
Titanic_df = Titanic[['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch','Fare']]
Titanic_df
```

```python
Titanic_df= Titanic_df.drop('Sex',axis=1)
Titanic_df
```

```python
plt.figure(figsize=(8,6),dpi=20)
sns.clustermap(data=Titanic_df,cmap='crest',col_cluster=False,vmin=10,vmax=800)
```

```python
plt.figure(figsize=(10,8),dpi=150)
sns.histplot(data=Titanic,x='Age',bins=50,color='pink')
```

```python
plt.figure(figsize=(10,8),dpi=150)
sns.boxplot(data=Titanic,x='Survived',y='Fare')
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.scatterplot(data=Titanic,x='Age',y='Fare')
plt.title("Age vs Fare")
```

```python
sns.countplot(data=Titanic,x='Survived')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.title('Survival Count')
```
